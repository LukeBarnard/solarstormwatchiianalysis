```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import os
import SolarStormwatchIIAnalysis as ssw
import h5py
import hi_processing as hip
import sunpy.map as smap
import astropy.units as u
%matplotlib inline
```

```python
# Test the STEREO A Coordinate conversions:

for craft in ['sta', 'stb']:
    # Get the STA file
    times = "2012-01-27T00:09:01"
    t_start = pd.to_datetime(times)
    files = hip.find_hi_files(t_start, t_start, craft=craft, camera='hi1', background_type=1 )
    print times,files

    # Make a sunpy map 
    himap = smap.Map(files)

    # Get arrays of pixel coordinates and use Sunpy to convert to HPC
    x = np.arange(0,himap.dimensions.x.value)
    y = np.arange(0,himap.dimensions.y.value)
    xm, ym = np.meshgrid(x, y)
    lon, lat = himap.pixel_to_data(xm*u.pix, ym*u.pix)

    # Do my conversion to HPR coords
    el, pa = hip.convert_hpc_to_hpr(lon, lat)
    # Do my back conversion to HPC coords
    lon2, lat2 = hip.convert_hpr_to_hpc(el, pa)

    #Plot them all out
    fig, ax = plt.subplots(1, 3, figsize=(12, 7))
    # HPC from Sunpy
    
    lat_levels = [-5, 0, 5]
    el_levels = [5, 10 ,15]
    if craft == 'sta':
        lon_levels = [-15, -10, -5]
        pa_levels = [60, 90 , 120]
    elif craft =='stb':
        lon_levels = [5, 10, 15]
        pa_levels = [240, 270 , 300]
        
    ax[0].contour(xm, ym, lon, levels=lon_levels, colors=['r','m','y'])
    ax[0].contour(xm, ym, lat, levels=lat_levels, colors=['r','m','y'])
    ax[0].set_title(craft.upper() + " Sunpy HPC")
    # My conversion to HPR
    ax[1].contour(xm, ym, el, levels=el_levels, colors=['r','m','y'])
    ax[1].contour(xm, ym, pa, levels=pa_levels, colors=['r','m','y'])
    ax[1].set_title(craft + " converted to HPR")
    # My conversion back to HPC
    ax[2].contour(xm, ym, lon2, levels=lon_levels, colors=['r','m','y'])
    ax[2].contour(xm, ym, lat2, levels=lat_levels, colors=['r','m','y'])
    ax[2].set_title(craft.upper() + " converted back to HPC")
    for a in ax:
        a.set_aspect('equal')

    # Check the HPC from Sunpy and my too-from conversion are equal within 1e-6
    tol = 1e-6
    lon_test = np.all(np.isclose(lon.value, lon2.value, atol=tol))
    lat_test = np.all(np.isclose(lon.value, lon2.value, atol=tol))
    print "{0} test: All longitudes match? {1}".format(craft.upper(), lon_test)
    print "{0} test: All latitudes match? {1}".format(craft.upper(), lat_test)
```
